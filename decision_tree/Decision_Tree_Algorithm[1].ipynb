# Finding Root Node

```python
def findInfo(valueDict, col, key):
    countY = 0
    for i in range(len(df)):
        if df[col][i] == key:
            if df['buys_computer'][i] == 'yes':
                countY += 1
    return (-(countY/valueDict[key]) * math.log(countY/valueDict[key] , 2) -((valueDict[key] - countY)/valueDict[key]) * math.log((valueDict[key] - countY)/valueDict[key] , 2))

column = df.columns.tolist()
column.pop()
root = ''
maxGain = -1
for col in column:
    valueDict = {}
    for row in range(len(df)):
        if df[col][row] not in valueDict:
            valueDict[df[col][row]] = 1
        else:
            valueDict[df[col][row]] += 1
        
    info = 0
    for key in valueDict.keys():
        info += round((valueDict[key]/len(df))*findInfo(valueDict, col,key) , 3) 
    gain = round(InfoD - info, 3)
    print(col," ", '->'," ",gain)
    if gain > maxGain:
        maxGain = gain
        root = col
```
